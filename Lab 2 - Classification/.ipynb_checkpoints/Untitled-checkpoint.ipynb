```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
exp1 = [pd.read_csv("exp1a.csv"),pd.read_csv("exp1b.csv"),pd.read_csv("exp1c.csv")]
# y.ravel()
```

```python
def NNc(data):
    X = data[["x", "y"]]
    y = data[["class"]]
    y = y['class']
    k_values = [1, 5, 11, 21]
    
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        kf = KFold(n_splits=10)
        cv_scores = cross_val_score(knn,X ,y , cv=kf, scoring='accuracy')
        # Record and print the accuracy for this k
        print(f'K = {k}')
        print(f'10-fold Cross-Validation Accuracy: {np.mean(cv_scores):.4f}')
        
        # Compute and print the confusion matrix
        y_pred = cross_val_predict(knn, X, y, cv=kf)
        cm = confusion_matrix(y, y_pred)
        print('Confusion Matrix:')
        print(cm)
        print('-' * 40)
```

```python
for data in exp1:
    NNc(data)
```

```python
def plot(data):
    # Split the data into two classes
    class_1 = data[data['class'] == 1]
    class_minus_1 = data[data['class'] == -1]
    
    # Create a scatter plot with two classes
    plt.scatter(class_1['x'], class_1['y'], label='Class 1', marker='o')
    plt.scatter(class_minus_1['x'], class_minus_1['y'], label='Class -1', marker='x')
    
    plt.title('Scatter Plot with Two Classes (1 and -1)')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    
    # Show the plot
    plt.show()
```

```python
for data in exp1:
    plot(data)    
```
